```python
import unittest
from src.stat_engine import StatEngine

class TestStatEngine(unittest.TestCase):

    def test_mean_median_odd_even(self):
        engine_odd = StatEngine([1, 3, 5])
        self.assertEqual(engine_odd.get_mean(), 3)
        self.assertEqual(engine_odd.get_median(), 3)
        engine_even = StatEngine([1, 2, 3, 4])
        self.assertEqual(engine_even.get_median(), 2.5)

    def test_mode_multimodal_unique(self):
        self.assertEqual(StatEngine([1, 1, 2, 2]).get_mode(), [1, 2])
        self.assertEqual(StatEngine([1, 2, 3]).get_mode(), "All values are unique")

    def test_empty_list(self):
        with self.assertRaises(ValueError):
            StatEngine([])

    def test_std_variance(self):
        engine = StatEngine([1, 2, 3, 4, 5])
        self.assertAlmostEqual(engine.get_standard_deviation(), 1.5811, places=4)
        self.assertAlmostEqual(engine.get_variance(), 2.5, places=4)

unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestStatEngine))
```
