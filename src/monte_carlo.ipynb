```python
import random

def simulate_crashes(days: int, crash_prob: float = 0.045) -> float:
    if days <= 0:
        raise ValueError("Days must be a positive integer")
    crashes = sum(1 for _ in range(days) if random.random() < crash_prob)
    return crashes / days

if __name__ == "__main__":
    for day_count in [30, 365, 10000]:
        prob = simulate_crashes(day_count)
        print(f"Simulated probability over {day_count} days: {prob:.4f}")
```
