```python
from typing import List, Union
import math
from collections import Counter

class StatEngine:
    def __init__(self, data: Union[List, tuple]):
        if not isinstance(data, (list, tuple)):
            raise TypeError("Input must be a list or tuple.")

        if len(data) == 0:
            raise ValueError("Input data cannot be empty.")

        self.data = self._clean_data(data)

    
    # Data Cleaning (separated)

    def clean_data(self, data: Union[List, tuple]) -> List[float]:
        cleaned_data = []

        for x in data:
            if isinstance(x, (int, float)):
                cleaned_data.append(x)

            elif isinstance(x, str):
                try:
                    cleaned_data.append(float(x))
                except ValueError:
                    raise TypeError(f"Invalid string value: '{x}' cannot be converted to number.")

            elif x is None:
                raise TypeError("NoneType value detected in data.")

            else:
                raise TypeError(f"Unsupported data type: {type(x)}")

        if len(cleaned_data) == 0:
            raise ValueError("No valid numeric data after cleaning.")

        return cleaned_data

 
    # Central Tendency
 
    def get_mean(self) -> float:
        return sum(self.data) / len(self.data)

    def get_median(self) -> float:
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        mid = n // 2

        if n % 2 == 0:
            return (sorted_data[mid - 1] + sorted_data[mid]) / 2
        return sorted_data[mid]

    def get_mode(self):
        counts = Counter(self.data)
        max_freq = max(counts.values())

        if max_freq == 1:
            return "All values are unique"

        modes = [k for k, v in counts.items() if v == max_freq]
        return modes if len(modes) > 1 else modes[0]


    # Dispersion

    def get_variance(self, is_sample: bool = True) -> float:
        n = len(self.data)

        if is_sample and n < 2:
            raise ValueError("Sample variance requires at least 2 data points.")

        mean = self.get_mean()
        var_sum = sum((x - mean) ** 2 for x in self.data)

        return var_sum / (n - 1) if is_sample else var_sum / n

    def get_standard_deviation(self, is_sample: bool = True) -> float:
        return math.sqrt(self.get_variance(is_sample))


    # Outliers

    def get_outliers(self, threshold: float = 2):
        mean = self.get_mean()
        std = self.get_standard_deviation()

        return [x for x in self.data if abs(x - mean) > threshold * std]
```
